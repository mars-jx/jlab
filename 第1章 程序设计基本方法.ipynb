# 程序语言概述

<font color=DarkRed size=3.5 face="微软雅黑">  程序设计语言</font><font color=black size=3.5 face="微软雅黑">是计算机能够理解用户操作意图的一种交互体系，是按照一定规则组织的计算机指令，能够让计算机自动执行各种运算处理。</font><font color=Darkblue size=3.5 face="微软雅黑">程序设计语言分为三大类：</font><br>

<td bgcolor=orange><font size=3.5>(1) 机器语言：二进制代码（0和1）,直接操作计算机硬件</font></td>

<td bgcolor=orange><font size=3.5>(2) 汇编语言：使用助记符与机器语言中的指令对应。如，ADD D1 D3</font></td>

<td bgcolor=orange><font size=3.5>(3) 高级语言：高级语言更接近于自然语言。如，result=1+2</font></td>

<font color=blue size=3.5 face="微软雅黑">**高级语言按照计算机执行方式分为两类：静态语言和脚本语言。</font>

<td bgcolor=BurlyWood><font size=3.5>**静态语言采用编译执行；脚本语言采用解释执行。</font></td>

## Python语言的特点

<font color=DarkRed size=3.5 face="微软雅黑">  Python语言</font><font color=black size=3.5 face="微软雅黑">作为一中广泛使用的通用高级脚本语言，与其他语言相比，具有很多优点。</font><font color=Darkblue size=3.5 face="微软雅黑">程序设计语言分为三大类：</font><br>

<td bgcolor=orange><font size=3.5>(1) 语法简洁：代码行数相当于其他语言的1/10~1/5。</font></td>

<td bgcolor=orange><font size=3.5>(2) 平台独立：安装Python解释器的计算机均可执行。</font></td>

<td bgcolor=orange><font size=3.5>(3) 粘性扩展：对应不同的需求有相应的接口和函数库进行“粘连”。</font></td>


<td bgcolor=orange><font size=3.5>(4) 开源理念：充分发挥“集体智慧”的作用，为语言的发展奠定了良好的基础。</font></td>

<td bgcolor=orange><font size=3.5>(5) 通用灵活：应用于不同领域，如数据处理、科学计算、人工智能...</font></td>

<td bgcolor=orange><font size=3.5>(6) 强制可读：严格缩进的语法结构类似于人类书写，强化语言的可读性。</font></td>

<td bgcolor=orange><font size=3.5>(7) 支持中文：utf-8编码的使用。</font></td>

<td bgcolor=orange><font size=3.5>(8) 模式多样：语法结构支持面向对象和面向结构，编程模式灵活。</font></td>

<td bgcolor=orange><font size=3.5>(9) 类库丰富：内置函数库+第三方库。</font></td>

## Python语言的应用

# 程序的基本编写方法

## IPO程序编写方法

<font color=DarkRed size=3.5 face="微软雅黑">  IPO模式（Input，Process，Output）</font><font color=black size=3.5 face="微软雅黑">是适用于任何程序语言编写的朴素运算模式。</font>

<td bgcolor=orange><font size=3>** 输入（Input）：程序的开始。输入可以是文件输入、网络输入、内部参数输入等。</font></td>

<td bgcolor=orange><font size=3>** 处理（Process）：程序对输入数据进行计算产生结果的过程，处理统称为“算法”。</font></td>

<td bgcolor=orange><font size=3>** 输出（Output）：展示程序运行结果。有多种形式，如文件输出、网络输出等。</font></td>

```python
###例1 ： 计算矩形面积

## 输入：矩形的长和宽
l = 10
w = 6

## 处理：面积 = 长 * 宽
area = l * w

## 输出：矩形的面积
print ('长为{}、宽为{}的矩形面积为：'.format(l,w),area)
```
