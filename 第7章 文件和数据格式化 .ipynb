# 文件的使用

<font color=blue size=3.5 face="微软雅黑">文件是存储在辅助存储器上的一组数据序列，可以包含任何数据内容。<br><br>概念上，文件是数据的集合和抽象。文件包括两种类型：<mark>文本文件和二进制文件</mark>。</font>

## 文件的类型

<font color=blue size=3.5 face="微软雅黑">文本文件一般由单一特定编码的字符组成，如UTF-8编码，内容容易统一展示和阅读。</font>

<font color=blue size=3.5 face="微软雅黑">二进制文件直接由比特0和比特1组成，文件内部数据的组织格式与文件用途有关。二进制是信息按照非字符但特定格式形成的文件，例如，png格式的图片文件、avi格式的视频文件。</font>

<font color=darkbrown size=3.5 face="微软雅黑">二进制文件和文本文件最主要的区别在于是否有统一的字符编码。</font>

<font color=darkbrown size=3.5 face="微软雅黑">无论文件创建为文本文件或者二进制文件，都可以用“文本文件方式”和“二进制文件方式”打开，但打开后的操作不同。</font>

## 文件的打开和关闭

<font color=blue size=3.5 face="微软雅黑">Python对文本文件和二进制文件采用统一的操作步骤，即“<mark>打开-操作-关闭</mark>”。</font>

<font color=darkbrown size=3.5 face="微软雅黑">(1) Python通过open()函数打开一个文件，并返回一个操作这个文件的变量，语法形式如下：</font>

<font color=darkorange size=3.5 face="微软雅黑"><变量名> = open(<文件路径及文件名>, <打开模式>)</font>

<font color=darkbrown size=3.5 face="微软雅黑">(2) 打开模式使用字符串方式表示，根据字符串定义，单引号或者双引号均可。上述打开模式中，'r'、
'w'、'x'、'b'可以和'b'、't'、'+'组合使用，形成既表达读写又表达文件模式的方式。</font>

<font color=darkbrown size=3.5 face="微软雅黑">(3) 文件使用结束后要用close()方法关闭，释放文件的使用授权，语法形式如下：</font>

<font color=darkorange size=3.5 face="微软雅黑"><变量名>.close()</font>

```python
path = 'G:\\jupyter\\'
fr = open(path + 'test.txt','rt')
print (fr.readline())
print (fr.readlines())
fr.close()
```

```python
fr = open(path + 'test.txt','rb')
print (fr.readline())
print (fr.readlines())
fr.close()
```

# 数据的组织维度

<font color=blue size=3.5 face="微软雅黑">一组数据在被计算机处理前需要进行一定的组织，表明数据之间的基本关系和逻辑，进而形成“数据的维度”。根据数据的关系不同，数据组织可以分为：<font color=red size=3.5 face="微软雅黑">一维数据、二维数据和高维数据。</font></font>

## 一维数据

<font color=blue size=3.5 face="微软雅黑" line-height=10>一维数据由对等关系的有序或无序数据构成，采用<font color=red size=3.5 face="微软雅黑">线性方式组织</font>，对应于数学中数组的概念。<br><br>例如：中国的直辖市列表即可表示为一维数据，一维数据具有线性特点。</font>

```python
temp = '''北京、上海、天津、重庆'''
tl = temp.split('、')
print(tl)
```

## 二维数据

<font color=blue size=3.5 face="微软雅黑">二维数据，也称表格数据，由关联关系数据构成，采用<font color=red size=3.5 face="微软雅黑">二维表格</font>方式组织，对应于数学中的矩阵，常见的表格都属于二维数据。<br><br>
例如：国家统计局发布的居民消费价格指数是二维数据。</font>

```python
zb = ['price','food','tobacco']
year = ['2014','2015','2016']
value = [[102,101.4,102],[103.1,102.3,104.6],[99.4,102.1,101.5]]
for i in value:
    print (i)
```

## 高维数据

<font color=blue size=3.5 face="微软雅黑">高维数据由键值对类型的数据构成，采用对象方式组织，可以多层嵌套。<br><br>高维数据在Web系统中十分常用，作为当今 Internet组织内容的主要方式，高位数据衍生出 HTML、XML、JSON等具体数据组织的语法结构。</font>

```python
dic = {'name':'lp','sex':'m','phone-number':['12312312','457457457654']}
print(dic)
```

# 二维数据的处理

## 二维数据的表示

<font color=blue size=3.5 face="微软雅黑">二维数据由多条一维数据构成，可以看成是一维数据的组合形式。因此，二维数据可以采用二维列表来表示。 </font>

```python
ls = [ 
['指标', '2014年', '2015年', '2016年'], ['居民消费价格指数', '102', '101.4', '102'], ['食品', '103.1', '102.3', '104.6'], ['烟酒及用品', '994', '102.1', '101.5'], ['衣着', '102.4', '102.7', '101.4'], ['家庭设备用品', '101.2', '101', '100.5'], ['医疗保健和个人用品', '101.3', '102', '101.1'], ['交通和通信', '99.9', '98.3', '98.7'], ['娱乐教育文化', '101.9', '101.4', '101.6'], ['居住', '102', '100.7', '101.6'], 
]
for i in ls:
    print (i)
```

## 二维数据的存储

<font color=blue size=3.5 face="微软雅黑">二维数据由一维数据组成，用CSV格式文件存储。存储方法：遍历+join方法。</font>

```python
fw = open(r'D:\Jupyter\python学习\pic\test_fw.csv','w')
for i in ls:
    fw.write(','.join(i) + '\n')
fw.close()
```

## 二维数据的处理

<font color=blue size=3.5 face="微软雅黑">对二维数据进行处理首先需要从CSV格式文件读入二维数据，并将其表示为二维列表对象。</font>

```python
fr = open(r'D:\Jupyter\python学习\pic\test_fw.csv','r')
tl = []
for i in fr:
    temp = i.strip().split(',')
    tl.append(temp)
fr.close()

for i in tl:
    print(i)
```

```python
for item in tl:
    line = ''
    for i in item:
        print ('{:^15}'.format(i), end='\t')
        
    print (line)
```
