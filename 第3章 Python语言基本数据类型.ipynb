## format()方法及使用 

### 基本使用

<font size=3.5 color=blue face=“微软雅黑”>字符串整合变量时需要使用字符串格式化方法，即字符串和变量同时输出时需要格式化。Python中推荐使用format（）格式化方法，具体格式为。</font>

<td bgcolor=orange><font size=3 face="微软雅黑"> <字符串常量或字符串变量\>.format(参数)</font></td>

<td bgcolor=orange><font size=3 face="微软雅黑"> 需要匹配的位置用{}标识。</font></td>

```python
s = '{}曰，有朋自远方{}，不亦{}乎'
print(s)
print (s.format('子','来','乐'))
```

```python
s = '{1}曰，有朋自远方{0}，不亦{2}乎'
print (s.format('乐','子','来'))
```

<font size=3.5 color=red face=“微软雅黑”>推荐使用第一种方法。</font>

### 格式控制

<font size=3.5 color=blue face=“微软雅黑”>format（）除了序号之外，也可以使用格式控制信息。</font>

<td bgcolor=orange><font size=3 face="微软雅黑"> {<参数序号\>:<格式控制标记\>}</font></td>

<td bgcolor=orange><font size=3 face="微软雅黑"> <、\>、^分别表示左对齐、右对齐和居中。</font></td>

```python
s = '好好学习'
print(s)
print ('{:5}'.format(s))
print ('{:^25}'.format(s))
print ('{:<2}'.format(s))
print ('{:>2}'.format(s))
```

<td bgcolor=orange><font size=3 face="微软雅黑"> {待格式化字符串：{填充字符串}对其标识 长度}</font></td>

```python
y = '#'#
z = '*'
print ('{0:{2}^25}'.format(s,y,z))
print ('{0:{1}>25}'.format(s,y))
print ('{0:{1}<25}'.format(s,y))
```

<td bgcolor=orange><font size=3 face="微软雅黑"> {带格式化数值：，（或.）类型}，其中“，”表示千位分割，“.”表示精度</font></td>

```python
v = 1234567890
print('{0:{1}^25,}'.format(v,y))
print('{0:*^25}'.format(v))

w = 12345.6789
print ('{0:{1}^25.3f}'.format(v,y))
print ('{0:*>30.2f}'.format(w))
```

<td bgcolor=orange><font size=3 face="微软雅黑"> 整数类型，输出格式6种：</font></td>

- b：整数的二进制

- c：整数的Unicode字符

- d：整数十进制

- o：整数八进制

- x：整数十六进制（小写）

- X：整数十六进制（大写）

```python
print("{:b}".format(147),"{:c}".format(147),      "{:d}".format(147),"{:o}".format(147),
      "{:x}".format(147),"{:X}".format(147))
```

<td bgcolor=orange><font size=3 face="微软雅黑"> 浮点数，输出格式4种：</font></td>

- e：对应e的指数形成

- E：对应E的指数形成

- f：对应标准形式

- %：对应百分比形式

```python
f = 3.1415926
print(type('{:e}'.format(f)),type(f)) #'{:e}'.format(f)
print ('{:e}'.format(f),'{:E}'.format(f),'{:f}'.format(f),'{:%}'.format(f))
```
