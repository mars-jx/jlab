## Python程序格式框架

<td bgcolor=orange><font size=3 face="微软雅黑">Python语言通过<mark>严格缩进</mark>来表明程序的格式框架。</font></td>

<td bgcolor=orange><font size=3 face="微软雅黑"> 严格缩进指的是每行代码开始前的空白。</font></td>

```python
#例1 九九乘法表
for i in range(1,10):
    for j in range(1,i+1):
        print (i * j,end = ' ')
    print ('')


print('打印结束')
```

# 命名与保留字

## 变量与命名 

<font size=3 color=blue face="微软雅黑"> 与数学概念类似，Python程序采用<mark>“变量”</mark>来保存和表示具体的数据值，相应变量需要关联一个标识符（即给变量起名字），由此，变量关联标识符的过程称为变量<font color=darkred>命名</font>。变量合法的命名规则如下。</font>

<td bgcolor=orange><font size=3 face="微软雅黑"> （1）大小写字母、数字、下划线和汉字等字符及其组合。</font></td> 

<td bgcolor=orange><font size=3 face="微软雅黑"> （2）首字符不能为数字，中间不能出现空格，长度不限。</font></td> 

<td bgcolor=orange><font size=3 face="微软雅黑"> ** 示例：\_x，x1，s\_x，x\_，xyz。</font></td> 

```python
_x = 1
x1 = 2
s_x = 3
x_ = 4
xyz = 5
print (_x,x1,s_x,x_,xyz)
```

# 函数

## 函数的概念 

<font size=3 color=blue face="微软雅黑"> 在程序编写过程中，一些特定的功能往往要反复使用，为了程序的编写便捷，将这些特定功能封装在一起，称为<font color=darkred>函数</font>。Python语言中函数的概念与数学中函数的概念类似。</font>

<td bgcolor=orange><font size=3 face="微软雅黑">（1）分类：内置函数、第三方库函数、自定义函数。</font></td> 

<td bgcolor=orange><font size=3 face="微软雅黑">（2）第三方库函数的使用，需要调用库。</font></td> 

<font size=2 color=darkred face="微软雅黑"> - 调用语句为：<br><br> <mark>import 库名 / import 库名 as 简名</mark> <br><br> <mark>from 库名 import *(函数名) / import 库名.函数名</mark></font>

<font size=2 color=darkred face="微软雅黑"> - 调用后具体使用：<br><br> <mark>库名.函数名(函数参数)</mark> <br><br> <mark>函数名(函数参数)</mark></font>


```python
import time as t #调用time中的time()函数

limit = 10 * 1000 * 1000

start = t.time()
while True:
    limit -= 1
    if limit <= 0:
        break

delta = t.time() - start
print('程序运行时间是：{}秒'.format(delta))
```

```python
import turtle as tl

tl.color('red', 'red')
tl.begin_fill()
for i in range(5):
    tl.fd(200)
    tl.speed(1)
    tl.rt(144)
tl.end_fill()
tl.done()
```

```python
from turtle import *
from random import *
def drawSnow():
    hideturtle()
    pensize(3)
    for i in range(19):
        r,g,b = random(),random(),random()
        pencolor(r,g,b)
        penup()
        setx(randint(-350,350))
        sety(randint(1,270))
        pendown()
        dens = randint(8,12)
        snowsize = randint(8,12)
        for j in range(1,dens):
            forward(snowsize)
            backward(snowsize)
            right(360/dens)

def drawGround():
    ht()
    for i in range(400):
        pensize(randint(5,10))
        x = randint(-400,350)
        y = randint(-280,-1)
        r,g,b = -y/280,-y/280,-y/280
        pencolor((r,g,b))
        penup()
        goto(x,y)
        pendown()
        fd(randint(40,100))
            
setup(800,600,200,200)
tracer(False)
bgcolor('black')
drawSnow()
drawGround()
done()
```
