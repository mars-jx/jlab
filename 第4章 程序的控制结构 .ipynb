# 程序的分支结构 

## 单分支结构：if 

<font size=3.5 color=blue face=“微软雅黑”>Python单分支结构使用if关键字来进行条件判断，格式如下。</font>

<td bgcolor=orange><font size=3 face="微软雅黑"> if <条件>：<br> <语句块></font></td>

<font size=3.5 color=darkbrown face=“微软雅黑”><语句块>是满足if条件后，程序执行的一条或多条语句。<条件>是一个产生True或False的语句，当结果为True时，<语句块>执行；当结果为False时，<语句块>不执行。</font>

```python
#例1：判断输入数字的奇偶性
num = eval(input('请输入一个整数：'))
if num % 2 != 0:
    print('{}是一个偶数。'.format(num))
```

```python
#例2： 闰年的判断
num = eval(input('请输入一个年份：'))
if (num % 4 == 0 and num % 100 != 0) or (num % 400 == 0):
    print('{}是闰年。'.format(num))
else:
    print('{}不是闰年。'.format(num))
        
```

## 二分支结构：if-else 

<font size=3.5 color=blue face=“微软雅黑”>Python二分支结构使用if-else关键字来进行条件判断，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> if <条件>：<br> &ensp;&ensp;<语句块1><br> else:<br>&ensp;&ensp;<语句块2></font>

```python
#例1：判断输入数字的奇偶性程序的修改

num = eval(input('请输入一个整数：'))
if num % 2 == 0:
    print('{}是一个偶数。'.format(num))
else:
    print('{}是一个奇数。'.format(num))
```

```python
#例2： 闰年的判断的修改

num = eval(input('请输入一个年份：'))
if (num % 4 == 0 and num % 100 != 0) or (num % 400 == 0):
    print('{}是闰年。'.format(num))
else:
    print('{}不是闰年。'.format(num))
```

## 多分支结构：if-elif-else

<font size=3.5 color=blue face=“微软雅黑”>Python多分支结构使用if-elif-else关键字来进行条件判断，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> if <条件1>：<br> &ensp;&ensp;<语句块1><br> elif <条件2>:<br>&ensp;&ensp;<语句块2><br>...<br> else:<br>&ensp;&ensp;<语句块3></font>

<font size=3.5 color=red face=“微软雅黑”>注意：使用if-elif-else进行条件判断时，要注意多个条件的先后关系。</font>

```python
# 例1：百分制转化为五级分制

score = eval(input('输入一个百分制分数：'))
grade = ''
if score >= 60:
    grade = 'D'
elif score >= 70:
    grade = 'C'
elif score >= 80:
    grade = 'B'
elif score >= 90:
    grade = 'A'
else:
    grade = 'E'
print ('对应的五级分制是：',grade)
    
```

```python
score = eval(input('输入一个百分制分数：'))
grade = ''
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'E'
print ('对应的五级分制是：',grade)
```

## 遍历循环for

<font size=3.5 color=blue face=“微软雅黑”>（1）Python遍历循环结构使用for关键字来实现，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> for <循环变量> in <遍历结构>：<br><br> &ensp;&ensp;&ensp; <语句块></font>

```python
x = ''
for i in 'Python':
    print(i)
    x = i
#print(i)
```

<font size=3.5 color=blue face=“微软雅黑”>（2）使用range()函数，可以指定循环次数，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> for <循环变量> in range(<循环次数>)：<br><br> &ensp;&ensp;&ensp; <语句块1></font>

<font size=3.5 color=blue face=“微软雅黑”>（3）遍历循环结构的扩展，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> for <循环变量> in <遍历结构>：<br><br> &ensp;&ensp;&ensp; <语句块1><br><br>else：<br><br>&ensp;&ensp;&ensp;<语句块2></font>

```python
for i in 'Python':
    print (i)
else:
    print ('循环结束')
```

-捕捉键盘输入的字符串，统计其中数字出现的次数。

```python
n = '0123456789'
count = 0
s = input('string:')
for c in s:
    if c in n:
        count = count + 1
print(count)
```

## 无限循环：while

<font size=3.5 color=blue face=“微软雅黑”> (1) Python语言通过保留字while实现无限循环，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> while <条件> ：<br><br> &ensp;&ensp;&ensp; <语句块><br><br></font>

<font size=3.5 color=darkbrown face=“微软雅黑”>注意：<条件>与if中的条件一样，返回True或False。</font>

```python
# 例1 ： 1-10求和

n = 0
s = 0
while n<=10:
    s = s + n
    #n = n + 1 
    n += 2
    print ('n=',n,' s=',s)
```

```python
s = 0
for i in range(0,11,1):
    s = s + i
    #s += i
print(s)
```

<font size=3.5 color=blue face=“微软雅黑”> (2)while无限循环也有扩展模式，与遍历循环的扩展模式类似，格式如下。</font>

<font size=3 color=darkorange face="微软雅黑"> while <条件>：<br><br> &ensp;&ensp;&ensp; <语句块1><br><br>else：<br><br>&ensp;&ensp;&ensp;<语句块2></font>

```python
n = 1
s = 0
while n<=10:
    s = s + n
    n = n + 1
else:
    print ('循环结束！')
print (n,s)
```

## 循环控制：break 和 continue 

<font size=3.5 color=blue face=“微软雅黑”> Python有两个控制循环结构的保留字，break和continue。<br>
<br>
-- break 用来跳出最内层for或while循环，脱离循环后执行循环之后的语句。<br>
<br>
-- continue 用来跳出当次循环，即当次循环后面的语句不再执行，而不跳出当前循环。</font>

```python
# 例1
while 1 == 1: #True
    s = input('输入一个字符：')
    if s == 'q' or s == 'Q':
        break
print ('循环结束')
```

```python
# 例2

s = 'Python'
for i in s:
    if i == 'o':
        break
    print (i)
print('循环结束')
```

```python
#例3

while 1 == 1: #True
    s = input('输入一个字符：')
    if s == 'q' or s == 'Q':
        break
    for i in s:
        if i == 'o':
            break
        print (i)
print ('循环结束')
```

```python
for i in range(2):
    for j in range(2):
        print('T')
```

# 程序的异常处理

<font size=3.5 color=blue face=“微软雅黑”> Python程序往往对输入有特定要求，当输入错误时会造成程序终止，即报错。<br>
<br>
为了保证程序运行的稳定性，Python语言使用try-except关键字实现异常处理。<br>
</font>

```python
try:
    num = eval(input('请输入一个自然数：'))
    tt = str(num)
    n = ''
    i = -1
    while i >= -len(tt):
        n = n + tt[i]
        i = i - 1
    print(n)
    if n == tt:
        print('回文数')
except:
    print('由于某些原因，出错了')
```

<font size=3.5 color=darkbrown face=“微软雅黑”>输入一行字符，统计不同类型字符的个数，字符类型包括英文字母、数字、其他类型字符。提示：大小写字母ASCII码分别从65和97开始，</font>

```python
t = '0123456789' 
num = 0
st = 0
sp = 0
other = 0
s = input('输入一行字符：')
for i in s:
    if (ord(i) >= 65 and ord(i) <= 90) or (ord(i) >= 97 and ord(i) <= 132):
        st = st + 1
    elif i in t:
        num = num + 1
    elif i == ' ':
        sp = sp + 1
    else:
        other = other + 1
print ('英文字符个数为：{}'.format(st),'\n','数字个数为：{}'.format(num),'\n','空格个数为：{}'.format(sp),'\n','其他字符个数为：{}'.format(other))
```
