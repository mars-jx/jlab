# 组合数据类型的基本概念

<font color=blue size=3.5 face="微软雅黑">通常情况下，程序不仅要处理单个数据，还要处理一组数据。<br><br>这种能够表示多个数据的类型称为<mark>组合数据类型</mark>。<br><br>Python中常用的组合数据类型有集合类型、序列类型和映射类型。</font>

## 集合类型概述

<font color=blue size=3.5 face="微软雅黑">（1）集合类型是一个或多个元素的集合，元素间无序，且元素唯一。</font>

<font color=blue size=3.5 face="微软雅黑">（2）Python语言中集合类型与数学中的集合概念一致，用大括号（{}）标识，无索引概念，可以动态添加或删除。</font>

<font color=blue size=3.5 face="微软雅黑">（3）集合类型中的元素类型必须是不可变数据类型，如整数、字符串。后面学习的列表、字典、以及集合本身都属于可变数据类型，不能作为集合的元素。</font>

```python
s1 = {100,'100',0.5}
s2 = set()
s3 = {'',}
s2.add(3)

print(type(s1),type(s2),type(s3))
print(s1,s2,s3)
```

<font color=darkbrown size=3. face="微软雅黑">注意：（1）由于集合元素的无序性，定义的顺序和输出顺序可能不一致。<br><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
    （2）由于集合元素的唯一性，可以使用集合去重。</font>

```python
s = {100,100,100}
print (s)
```

# 字典类型

<font color=blue size=3.5 face="微软雅黑">“键值对”是组织数据的一种重要方式，广泛应用在Web系统中。键值对的基本思想是将“值”信息关联一个“键”信息，进而通过键信息查找对应值信息，这个过程叫映射。Python语言中通过字典类型实现映射。</font>

## 字典的定义 

<font color=blue size=3.5 face="微软雅黑">Python语言中的字典使用大括号{}建立，每个元素是一个键值对，使用方式如下。</font>

<font color=darkorange size=3.5 face="微软雅黑">{<键1>:<值1>, <键2>:<值2>, … , <键n>:<值n>}</font>

<font color=darkbrown size=3.5 face="微软雅黑">键和值通过冒号连接，不同键值对通过逗号隔开。字典类型也具有和集合类似的性质，即键值对之间没有顺序且不能重复。</font>

```python
dic = {'a':1,'b':2,'c':3}
print (dic)
```

```python
dic = dict()
print (dic)
```

```python
dic = {}
dic,type(dic)
```

## 字典的索引

<font color=blue size=3.5 face="微软雅黑"> (1)列表类型采用元素顺序的位置进行索引。字典元素“键值对”中键是值的索引，因此，可以直接利用键值对关系索引元素。</font>

<font color=darkorange size=3.5 face="微软雅黑"><值> = <字典变量>[<键>]</font>

```python
dic['a']
```

<font color=blue size=3.5 face="微软雅黑"> (2)利用索引和赋值（=）配合，可以对字典中每个元素进行修改。</font>

```python
dic['a'] = 5
dic
```

<font color=blue size=3.5 face="微软雅黑"> (3)使用大括号可以创建字典。 通过索引和赋值配合，可以向字典中增加元素。</font>

```python
dic = {}
dic['a'] = [1]
dic['b'] = 2
dic
```

<font color=darkbrown size=3.5 face="微软雅黑">字典是存储可变数量键值对的数据结构，键和值可以是任意数据类型，通过键索引值，并可以通过键修改值。</font>

# 字典类型的操作

## 字典的操作函数 

<font color=blue size=3.5 face="微软雅黑"> 字典类型有一些通用的操作函数。</font>

<font color=darkbrown size=3.5 face="微软雅黑">(1)len(d)给出字典d的元素个数，也称为长度。</font>

```python
len(dic)
```

<font color=darkbrown size=3.5 face="微软雅黑">(2)min(d)和max(d)分别返回字典d中最小或最大索引值。</font>

```python
dic = {1:100,2:90,3:85}
print (min(dic),max(dic))
```

## 字典的操作方法 

<font color=blue size=3.5 face="微软雅黑"> 字典类型存在一些操作方法，使用语法形式如下。</font>

<font color=darkorange size=3.5 face="微软雅黑"><字典变量>.<方法名称>(<方法参数>)</font>

<font color=darkbrown size=3.5 face="微软雅黑">(1)d.keys()返回字典中的所有键信息，返回结果是Python的一种内部数据类型dict_keys，专用于表示字典的键。如果希望更好的使用返回结果， 可以将其转换为列表类型。</font>

```python
temp = {'1001':90,'1002':80,'1003':65,'1004':91,'1005':70,'1006':92}
print (temp.keys())
for k in temp.keys():
    print(k)
```

<font color=darkbrown size=3.5 face="微软雅黑">(2)d.values()返回字典中的所有值信息，返回结果是Python的一种内部数据类型dict_values。如果希望更好的使用返回结果， 可以将其转换为列表类型。</font>

```python
print (temp.values())
print (list(temp.values()))
for v in temp.values():
    print(v)
```

<font color=darkbrown size=3.5 face="微软雅黑">(3)d.items()返回字典中的所有键值对信息，返回结果是Python的一种内部数据类型dict_items。</font>

```python
for item in temp.items():
    print(item)
```

<font color=darkbrown size=3.5 face="微软雅黑">(4)d.get(key, default)根据键信息查找并返回值信息，如果key存在则返回相应值，否则返回默认值， 第二个元素default可以省略， 如果省略则默认值为空。</font>

```python
print (temp.get('1001'))
print (temp.get('1007'))
print (temp['1001'])
print (temp['1007'])
```

<font color=darkbrown size=3.5 face="微软雅黑">(5)d.pop(key, default)根据键信息查找并取出值信息，如果key存在则返回相应值，否则返回默认值， 第二个元素default可以省略，如果省略则默认值为空。相比d.get()方法，d.pop()在取出相应值后，将从字典中删除对应的键值对。</font>

```python
temp['1007'] = 55
print (temp)
print(temp.pop('1007'))
print (temp)
```

<font color=darkbrown size=3.5 face="微软雅黑">(6)d.popitem()随机从字典中取出一个键值对，以元组(key,value)形式返回。取出后从字典中删除这个键值对。</font>

```python
print(temp.popitem())
print (temp)
```

<font color=darkbrown size=3.5 face="微软雅黑">(7)d.clear()删除字典中所有键值对。</font>

```python
a = {1:1,2:2}
a.clear()
print (a)
```

<font color=darkbrown size=3.5 face="微软雅黑">(8)如果希望删除字典中某一个元素，可以使用Python保留字del。</font>

```python
del temp['1001']
print (temp)
```

<font color=darkbrown size=3.5 face="微软雅黑">(9)字典类型也支持保留字in，用来判断一个键是否在字典中。如果在则返回True，否则返回False。</font>

```python
print ('1002' in temp, '1008' in temp)
```

<tr><td bgcolor=orange><font color=black size=3.5 face="微软雅黑">字典可以遍历循环对其元素进行
遍历：</font></td></tr> 

<font color=darkorange size=3.5 face="微软雅黑">for <变量名> in <字典名>: <br><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<语句块></font>

```python
for key in temp.keys():
    print (key,temp[key])
```

```python
for value in temp.values():
    print (value)
```

```python
for item in temp.items():
    print (item)
```

```python
dic = {'a':1,'b':[2,3,4]}
for item in dic.items():
    #print(item[0],item[1])
    if type(item[1]) == list:
        for i in item[1]:
            print(i)
    else:
        print(item[1])
```

```python
s = 'a b c a d b d'
lis = s.split(' ')
dic = {}
for i in lis:
    if i not in dic.keys():
        dic[i] = 1
    else:
        dic[i] = dic[i] + 1
dic
```
